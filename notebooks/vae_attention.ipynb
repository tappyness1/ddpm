```python
import torch
from torch import nn
```

```python
img_features = torch.randn(5, 32, 16, 16)
linear_layer = nn.Linear(32, 8*3)
n, c, h, w = img_features.size()
img_features_reshaped = img_features.view(n, c, h*w)
```

```python
out = linear_layer(img_features_reshaped.permute(0, 2, 1))
```

```python
out.chunk(3, dim=-1)
```
